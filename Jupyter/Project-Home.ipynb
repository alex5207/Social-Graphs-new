## 4. Hollywood as a sign of the times?
Following the analysis of the wiki data and the movie data separately, we now investigate whether there seems to be a correlation between them. The purpose is to explore whether the movie industry resembles the state of the real world, and thus act as a sign of the times.

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from IPython.display import display, Image
from PIL import Image
import numpy as np
```

### Wordclouds
Initially we present the movie- and wiki wordclouds next to one another, in order to compare them.

```python
def plotDecade(decade):
    
    movie_photo_path = '../assets/hollywood_tfidf_' + str(decade) + '.png'
    wiki_photo_path = '../assets/wiki_tfidf_' + str(decade) + '.png'
    
    img_movie = mpimg.imread(movie_photo_path)
    img_wiki = mpimg.imread(wiki_photo_path)

    # display images
    fig, ax = plt.subplots(1,2, figsize = (20, 12))
    ax[0].imshow(img_movie);
    ax[0].axis('off')
    ax[1].imshow(img_wiki);
    ax[1].axis('off')

    plt.show()
```

```python
for i in np.arange(1980, 2011, 10):
    plotDecade(i)
```

<b>No correlation</b>

Observing the wordclouds next to one another, it seems like Hollywood does not create movies that represent the state of the world. Therefore it does not seem that Hollywood movies and their focus is a good proxy for the situation of the real world - At least not in the sense that Wikipedia data represents the true state of the world (which is also questionable)

Before concluding whether Hollywood is a sign of the times or not, we want to compare the sentiment of Hollywood movies in some decades with the sentiment of the wikipages representing those decades.

### Sentiment

In this section we investigate whether the "mood" of hollywood movies resemble the mood of the decade, by comparing their sentiments.

```python
senti = mpimg.imread('../assets/sentiment_compar.png')

# display images
fig, ax = plt.subplots(1,1, figsize = (20, 15))
ax.imshow(senti);
ax.axis('off')

plt.show()
```

<b>Hard to analyze</b>

From the above plots, the sentinent seems rather constant throughout the decades, both with regards to what happens in the real world, and what happens in the movies. In the 1920's and the 2010's there seem to be some difference between the sentiment, where the 1920's are really happy days in the real world, it does not explode with happiness in movies. Also, the movies seem to be relatively sad in the 2010's, whereas the real world seem averagely happy.

However, the red bars indicate very large uncertainties, so it is arguably impossible to derive whether there is a correlation of the mood in the real world and the mood of the movies at some given time.

**From these results there does not seem to be any evidence that the movie industry in Hollywood is indeed a sign of the times.**
One could argue that we assign equal weight to all kinds of movies when doing this comparisson. An example can be that e.g. genres as fantasy, science fiction and action rarely care about the state of the real world, whereas political movies, dramas and documentaries often do. Therefore, there might be multiple movies resembling the state of the world, but because they are outweighted by movies that doesn't, we are not seeing this in our analysis.

### Limitations of the project
#### Looking at subset of graph
A large drawback of our network analysis is definitely the narrow scope of the project. Limiting ourselves to only looking at the very small subset of actors and disgarding a lot of low-degree nodes surely changed the network properties away from usual networks (e.g powerlaw distributed networks). Likely, this had a large effect on ability to genereate insights with community detection. Most of us already know that the cast of the "Oceans 11,12,13" movies had a lot to do with eachother. Working with the full network might have provided some unforseen and interesting observations.

#### Communities
With regards to the communities, it was observed that the detected communities were of poor quality. The problem could be the constraint of one actor is only able to in one community. By using link clustering, allowing actors to be in multiple communities, it could be the case that we would in fact get genre-based communities. It seems there simply is not enough evidence, to conclude that an actor's personal relationships to his/her co-stars, is the main driver.

#### Semantic change
Some words change their meaning. Silly, "in its earliest uses, it referred to things worthy or blessed; from there it came to refer to the weak and vulnerable, and more recently to those who are foolish" [\[7\]](https://ideas.ted.com/20-words-that-once-meant-something-very-different/). This is definitely something to be aware of when applying sentiment analysis over decades.
